```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#확률 전이 행렬 A

"""
확률 전이 행렬 -> 모든 열(세로)의 합이 1 (확률 보존 특성)
이로 인해 행렬 A의 고유값(Eigenvalue -> 방정식의 해) 중 하나는 반드시 1 (선형 대수학의 성질에 따름)
이 고유값 1에 대응하는 고유 벡터를 확률적으로 정규화하면 시스템이 수렴하게 되는 정상 상태 확률 분포가 됨
"""
A = np.array([[0.5,0.5,0.25],[0.25,0.0,0.25],[0.25,0.5,0.5]])
```

```python
print(A)
```

```python
xtoday = np.array([[1],[0],[0]]) #today의 상태 벡터 x
print(xtoday)
```

```python
print(A@xtoday)
```

```python
weather = np.zeros((3,50))
for k in range(50):
    xtomorrow = A@xtoday
    weather[:,k]=xtomorrow[:,0]
    xtoday = xtomorrow
    print("DAY : "+str(k+2))
    print(xtoday)
```

```python
plt.plot(weather.transpose())
plt.grid(True)
```
